    Comparison of 3 Techniques


**1. Neural Networks**


Strengths:
•	Pattern Recognition: Neural networks are used in machine learning where the desired patterns are formed by connecting nodes or neurons in layers.
•	Versatility: They perform fine in basic image categorization problems or wherever the dataset is less in size, or it is not complicated.
•	Scalability: The model is easily expandable by simply adding more layers or neurons in-terms of accuracy.
Weaknesses:
•	Limited Depth: It was also found that shallow networks may not be able to handle large, complex, high-dimensional data well.
•	Overfitting: High risk of memorization when working with small amounts of data if the methods of regularization are not used.
Advantages:
•	Ease of Use: Good for simple object detection without the need of computational complexity of other algorithms.
•	Interpretability: There are also relatively simpler to trace and analyse compared to deeper architectures.
Disadvantages:
•	Performance Limits: It performs worse in highly complex tasks than in the deeper networks.
•	Dependency on Feature Engineering: It often involves manual feature extraction which is not the case with CNNs.
Example Application: Recognizing basic geometric shapes such as circles, squares from a low resolution image database.
Input Data: Flattened pixel arrays.
Output: Group names (e.g., “Circle,” “Square”).


**2. Deep Learning (Deep Neural Networks)**


Strengths:
•	Automatic Feature Extraction: Does not require any human interaction to extract features of the data set.
•	High Accuracy: Outperforms in the identification of trends in large and unstructured data sets. Scalability: Applicable for multi-class and large scale object recognition problems.
Weaknesses:
•	Computationally Intensive: Very computationally intensive and time consuming.
•	Data Hungry: Requires large datasets and these datasets should be labeled.
Advantages:
•	Generalization: Obtains high level of generalization performance when adequate training data is available.
•	Flexibility: Flexible to other tasks, for example multi-object detection or segmentation.
Disadvantages:
•	Black Box Nature: Difficult to understand how the model comes to its conclusions.
•	Overfitting Risk: May overfit if not trained correctly or if the training data is insufficient.
Example Application: Recognizing more than one object in real-world street scenes (for example, automobiles and people).
Input Data: Image tensors of high resolution.
Output: Confidences or coordinates of the axis-aligned rectangles around each object class.


**3. Convolutional Neural Networks (CNNs)**


Strengths:
•	Localized Feature Learning: CNNs are good at identifying the spatial pyramid of images through convolution layers.  
•	Parameter Efficiency: Split weights between neurons, hence reducing the number of parameters.  
•	Robustness: It can handle noise, rotations and scaling very well.  
Weaknesses:
•	Complexity: Less easy to design and optimise than simple feed forward neural networks.  
•	Hardware Requirements: Needs GPUs for training and testing to be done effectively.  
Advantages:
•	Performance: Achieving the best results in the field of image recognition and object detection.  
•	Adaptability: Flexible enough to be easily modified for multi-object detection or instance segmentation.  
Disadvantages:
•	Specialization: Needs a lot of tuning for a certain task.  
•	Overhead: The training time and memory usage are more than simple models.  
Example Application: Identifying species of animals in high definition wildlife photographs.  
Input Data: Image tensors of spatial dimensions and colour channels.  
Output: Predictive classes or probabilities of occurrence.

```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import resnet50
from tensorflow.keras.utils import to_categorical

# Define paths
train_dir = '/content/drive/MyDrive/Object Recognition/Object Recognition/train'
test_dir = '/content/drive/MyDrive/Object Recognition/Object Recognition/test'

# Parameters
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32
CLASSES = ['bottles', 'pens', 'desks', 'shoes']  # Update this if needed

# Data augmentation for training data
train_datagen = ImageDataGenerator(
    rescale=1./255,                # Normalize pixel values
    rotation_range=20,             # Rotate images randomly
    width_shift_range=0.2,         # Translate images horizontally
    height_shift_range=0.2,        # Translate images vertically
    shear_range=0.15,              # Apply shear transformations
    zoom_range=0.15,               # Zoom into images randomly
    horizontal_flip=True,          # Flip images horizontally
    fill_mode='nearest'            # Fill empty pixels
)

# Only rescale for test data (no augmentation)
test_datagen = ImageDataGenerator(rescale=1./255)

# Create data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASSES
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASSES
)

# Check class indices mapping
print("Class indices mapping:", train_generator.class_indices)

# Example of loading and preprocessing a single batch
sample_batch = next(train_generator)
images, labels = sample_batch

print("Batch images shape:", images.shape)  # (BATCH_SIZE, IMG_HEIGHT, IMG_WIDTH, 3)
print("Batch labels shape:", labels.shape)  # (BATCH_SIZE, num_classes)

# Optional: Save preprocessed data for debugging or other purposes
import numpy as np

np.save('train_images.npy', images)
np.save('train_labels.npy', labels)

# Model-ready preprocessing function
def preprocess_input_fn(image_batch):
    return resnet50.preprocess_input(image_batch)  # Preprocess according to ResNet50

# Apply preprocessing to images
images_preprocessed = preprocess_input_fn(images)
print("Preprocessed images shape:", images_preprocessed.shape)
```

```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import resnet50
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping

# Paths
train_dir = '/content/drive/MyDrive/Object Recognition/Object Recognition/train'
test_dir = '/content/drive/MyDrive/Object Recognition/Object Recognition/test'

# Parameters
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32
CLASSES = ['bottles', 'pens', 'desks', 'shoes']

# Data augmentation for training data
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Only rescale for test data
test_datagen = ImageDataGenerator(rescale=1./255)

# Create data generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASSES
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASSES
)

# CNN Model
model = Sequential([
    Input(shape=(IMG_HEIGHT, IMG_WIDTH, 3)),
    Conv2D(32, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(train_generator.num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Early Stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = model.fit(
    train_generator,
    epochs=25,
    validation_data=test_generator,
    callbacks=[early_stopping]
)

# Evaluate on the test dataset
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Save the trained model
model.save('object_recognition_model.h5')
print("Model saved as 'object_recognition_model.h5'")

# Load and use the model later
from tensorflow.keras.models import load_model
loaded_model = load_model('object_recognition_model.h5')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Load the trained model
model = load_model('/content/drive/MyDrive/Object Recognition/object_recognition_model.h5')

# Class labels mapping (ensure this matches your dataset class indices)
class_indices = {'bottles': 0, 'pens': 1, 'desks': 2, 'shoes': 3}
classes = list(class_indices.keys())

# Function to predict a single image
def predict_image(image_path, model, classes):
    # Load and preprocess the image
    img = load_img(image_path, target_size=(224, 224))  # Resize to match training dimensions
    img_array = img_to_array(img) / 255.0  # Normalize pixel values
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

    # Make prediction
    prediction = model.predict(img_array)
    predicted_class_index = np.argmax(prediction)
    predicted_class_label = classes[predicted_class_index]
    confidence = prediction[0][predicted_class_index] * 100

    # Display the image and prediction
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Predicted: {predicted_class_label} ({confidence:.2f}%)")
    plt.show()

    return predicted_class_label, confidence

# Predict an image
image_path = '/content/drive/MyDrive/Object Recognition/Object Recognition/test/shoes/shoes (104).jpg'  # Replace with the path to your test image
predicted_label, confidence = predict_image(image_path, model, classes)

print(f"Predicted Class: {predicted_label}, Confidence: {confidence:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score
import seaborn as sns
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load the trained model
model = load_model('/content/drive/MyDrive/Object Recognition/object_recognition_model.h5')

# Class labels mapping (ensure this matches your dataset class indices)
class_indices = {'bottles': 0, 'pens': 1, 'desks': 2, 'shoes': 3}
classes = list(class_indices.keys())

# Directory containing the test dataset
test_dir = '/content/drive/MyDrive/Object Recognition/Object Recognition/test'  # Replace with the test dataset path

# Load test data using ImageDataGenerator
test_datagen = ImageDataGenerator(rescale=1./255)  # Normalize pixel values
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# Get true labels
true_labels = test_generator.classes

# Predict all images in the test set
predictions = model.predict(test_generator, verbose=1)
predicted_labels = np.argmax(predictions, axis=1)

# Confusion Matrix
conf_matrix = confusion_matrix(true_labels, predicted_labels)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report
class_report = classification_report(true_labels, predicted_labels, target_names=classes)
print("Classification Report:\n", class_report)

# Accuracy and F1-Score
accuracy = accuracy_score(true_labels, predicted_labels)
f1 = f1_score(true_labels, predicted_labels, average='weighted')

# Display Accuracy and F1-Score
print(f"Accuracy: {accuracy:.2f}")
print(f"F1-Score: {f1:.2f}")

# Accuracy and F1-Score Chart
metrics = {'Metric': ['Accuracy', 'F1-Score'], 'Value': [accuracy, f1]}
plt.bar(metrics['Metric'], metrics['Value'], color=['blue', 'orange'])
plt.ylim(0, 1)
plt.title("Model Evaluation Metrics")
plt.ylabel("Score")
for i, v in enumerate(metrics['Value']):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center')
plt.show()
```
